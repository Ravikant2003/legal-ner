```python
from transformers import AutoTokenizer, AutoModelForTokenClassification

model_name = "bert-base-uncased"  # or "nlpaueb/legal-bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from datasets import load_from_disk
train_dataset = load_from_disk("/kaggle/input/ner-mini/ner_dataset_encoded_Judgement_Train")
test_dataset = load_from_disk("/kaggle/input/ner-mini/ner_dataset_encoded_Judgement_Test")
eval_dataset = load_from_disk("/kaggle/input/ner-mini/ner_dataset_encoded_Judgement_Dev")
```

```python
import json
label2id = "/kaggle/input/ner-mini/label2id_Judgement_Train.json"
```

```python
with open("/kaggle/input/ner-mini/label2id_Judgement_Train.json") as f:
    label2id = json.load(f)

# Convert string keys back to integers, and reverse
id2label = {int(v): k for k, v in label2id.items()}
```

```python
model = AutoModelForTokenClassification.from_pretrained(
    model_name,
    num_labels=len(label2id),
    id2label=id2label,
    label2id=label2id
)
```

```python
from transformers import Trainer, TrainingArguments
from transformers import EarlyStoppingCallback

# Updated hyperparameters based on observations
training_args = TrainingArguments(
    output_dir="/kaggle/working/ner_model",
    eval_strategy="steps",
    eval_steps=500,  # Less frequent evaluation
    logging_steps=100,
    logging_dir="/kaggle/working/logs",
    report_to="none",
    run_name="ner_finetune_bert",
    disable_tqdm=False,
    per_device_train_batch_size=16,  # Keep it smaller, or increase to 32
    per_device_eval_batch_size=16,
    num_train_epochs=3,  # Keep it at 3, monitor performance
    save_steps=500,
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    weight_decay=0.01,
    learning_rate=1e-5,  # Reduce learning rate
    warmup_steps=300,  # Warmup steps, 10% of training steps
    gradient_accumulation_steps=2,  # Simulate larger batch size
  # Adjust for the length of your documents
)

# Early Stopping Callback with adjusted patience
early_stopping = EarlyStoppingCallback(early_stopping_patience=3)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    callbacks=[early_stopping]
)
```

```python
trainer.train()
trainer.evaluate()
```

```python
trainer.save_model("/kaggle/working/ner_model")
tokenizer.save_pretrained("/kaggle/working/ner_model")
```

```python
from transformers import AutoModelForTokenClassification, AutoTokenizer
import torch

model_path = "/kaggle/working/ner_model"

model = AutoModelForTokenClassification.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
import json

with open("/kaggle/input/ner-mini/label2id_Judgement_Train.json") as f:
    label2id = json.load(f)

id2label = {v: k for k, v in label2id.items()}
```

```python
from sklearn.metrics import classification_report

def evaluate_model(model, eval_dataset, id2label):
    model.eval()
    predictions = []
    references = []

    for sample in eval_dataset:
        input_ids = torch.tensor([sample["input_ids"]])
        attention_mask = torch.tensor([sample["attention_mask"]])

        #labels = sample["labels"]
        labels = torch.tensor(sample["labels"])

        with torch.no_grad():
            output = model(input_ids=input_ids, attention_mask=attention_mask)
        
        logits = output.logits
        preds = torch.argmax(logits, dim=2).squeeze().tolist()
        labels = labels.tolist()

        # Remove padding (-100) from both
        filtered_preds = [id2label[p] for p, l in zip(preds, labels) if l != -100]
        filtered_labels = [id2label[l] for l in labels if l != -100]

        predictions.extend(filtered_preds)
        references.extend(filtered_labels)

    print(classification_report(references, predictions, digits=4))

evaluate_model(model, eval_dataset, id2label)
```

```python
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    true_predictions = [
        [id2label[p] for (p, l) in zip(pred, lab) if l != -100]
        for pred, lab in zip(predictions, labels)
    ]
    true_labels = [
        [id2label[l] for (p, l) in zip(pred, lab) if l != -100]
        for pred, lab in zip(predictions, labels)
    ]

    precision, recall, f1, _ = precision_recall_fscore_support(
        list(chain(*true_labels)), list(chain(*true_predictions)), average="weighted"
    )
    return {
        "precision": precision,
        "recall": recall,
        "f1": f1,
    }
```
